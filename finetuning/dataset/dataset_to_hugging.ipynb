```python
from datasets import Dataset, Features, Value, Image
import pandas as pd
import os

# === CONFIG ===
csv_path = "/content/drive/MyDrive/200set/renamed_captions.csv"
image_dir = "/content/drive/MyDrive/200set/renamed_images"
hf_repo_name = "image_captions"  # ğŸ” change this if needed (username will be auto-added)

# === LOAD CSV ===
df = pd.read_csv(csv_path)

# === CREATE DATA LIST ===
data = []
for _, row in df.iterrows():
    img_filename = row["filename"]
    img_path = os.path.join(image_dir, img_filename)

    if not os.path.exists(img_path):
        print(f"âš ï¸ Skipping missing image: {img_path}")
        continue

    data.append({
        "image": img_path,  # the local path (datasets.Image will load it)
        "image_id": img_filename.split(".")[0],
        "caption": row["caption"].strip()
    })

# === CONVERT TO DATASET ===
dataset = Dataset.from_list(data).cast_column("image", Image())

# === OPTIONAL: PREVIEW ===
print(dataset[0])  # shows: image, image_id, caption

# === PUSH TO HUGGINGFACE HUB ===
from huggingface_hub import login
login(token="")  # âœ… already done earlier

# ğŸš€ Upload to your account (change visibility if you want)
dataset.push_to_hub("wolf010/" + hf_repo_name, private=False)
```
